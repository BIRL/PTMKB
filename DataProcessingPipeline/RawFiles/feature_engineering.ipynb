```python
import requests
def get_sequence(accession: str) -> str | None:
    try:
        response = requests.get(
            f'https://www.ebi.ac.uk/proteins/api/proteins/{accession}'
        )
        return response.json()['sequence']['sequence']
    except Exception as e:
        print("Couldn't get it because of", e)
        return None
```

```python
protein_sequence = get_sequence(accession)
```
